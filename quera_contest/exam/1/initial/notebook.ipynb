```python
# Regression imports
import pandas as pd
import numpy as np



from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, VotingRegressor
from xgboost import XGBRegressor

# Optional: visualization settings

```

```python
train = pd.read_csv("./Data/train.csv")
x_test = pd.read_csv("./Data/test.csv")
```

```python
train.isnull().sum()
```

```python
train["Cement Moisture Factor"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
corr = train.corr(numeric_only=True)

plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
train.drop(columns=["Cement Moisture Factor"],inplace=True,axis=1)
x_test.drop(columns=["Cement Moisture Factor"],inplace=True,axis=1)
```

```python
X = train.drop(columns=['Strength'])
y = train['Strength']
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
```

```python
model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(X_train_scaled, y_train)
preds = model.predict(X_val_scaled)
preds_train = model.predict(X_train_scaled)


print("R2:", r2_score(y_val, preds))
print("R2 train:", r2_score(y_train, preds_train))
```

```python
xgb_model = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)
xgb_model.fit(X_train_scaled, y_train)
xgb_preds = xgb_model.predict(X_val_scaled)
xgb_train_preds = xgb_model.predict(X_train_scaled)

print("XGBoost R2:", r2_score(y_val, xgb_preds))
print("XGBoost Train R2:", r2_score(y_train, xgb_train_preds))
print("—" * 40)
```
