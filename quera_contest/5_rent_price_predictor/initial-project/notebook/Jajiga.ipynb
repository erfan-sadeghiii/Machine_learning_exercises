```python
from sklearn.preprocessing import MultiLabelBinarizer

# Create separate binarizers
mlb_types = MultiLabelBinarizer()
mlb_regions = MultiLabelBinarizer()
mlb_rules = MultiLabelBinarizer()
mlb_features = MultiLabelBinarizer()

# Preprocess columns
for col in ["types", "regions", "rules"]:
    df[col] = df[col].map(eval)
    x_test[col] = x_test[col].map(eval)

df["features"] = df["features"].map(eval).map(lambda x: [e["name"] for e in x])
x_test["features"] = x_test["features"].map(eval).map(lambda x: [e["name"] for e in x])

# Fit and transform df
df = pd.concat([df.reset_index(drop=True),
                pd.DataFrame(mlb_types.fit_transform(df["types"]),
                             columns=[f'types_{c}' for c in mlb_types.classes_])], axis=1)

df = pd.concat([df.reset_index(drop=True),
                pd.DataFrame(mlb_regions.fit_transform(df["regions"]),
                             columns=[f'regions_{c}' for c in mlb_regions.classes_])], axis=1)

df = pd.concat([df.reset_index(drop=True),
                pd.DataFrame(mlb_rules.fit_transform(df["rules"]),
                             columns=[f'rules_{c}' for c in mlb_rules.classes_])], axis=1)

df = pd.concat([df.reset_index(drop=True),
                pd.DataFrame(mlb_features.fit_transform(df["features"]),
                             columns=[f'features_{c}' for c in mlb_features.classes_])], axis=1)

# Transform x_test using the same binarizers
x_test = pd.concat([x_test.reset_index(drop=True),
                    pd.DataFrame(mlb_types.transform(x_test["types"]),
                                 columns=[f'types_{c}' for c in mlb_types.classes_])], axis=1)

x_test = pd.concat([x_test.reset_index(drop=True),
                    pd.DataFrame(mlb_regions.transform(x_test["regions"]),
                                 columns=[f'regions_{c}' for c in mlb_regions.classes_])], axis=1)

x_test = pd.concat([x_test.reset_index(drop=True),
                    pd.DataFrame(mlb_rules.transform(x_test["rules"]),
                                 columns=[f'rules_{c}' for c in mlb_rules.classes_])], axis=1)

x_test = pd.concat([x_test.reset_index(drop=True),
                    pd.DataFrame(mlb_features.transform(x_test["features"]),
                                 columns=[f'features_{c}' for c in mlb_features.classes_])], axis=1)
```

```python
df["sleep_arrange"] = df["sleep_arrange"].map(lambda x: eval(x))
df["sleep_arrange_count"] = df["sleep_arrange"].map(lambda x: len(x))
df["sleep_arrange_single_sum"] = df["sleep_arrange"].map(lambda x: sum([e['single'] for e in x]))
df["sleep_arrange_double_sum"] = df["sleep_arrange"].map(lambda x: sum([e['double'] for e in x]))
df["sleep_arrange_mattress_sum"] = df["sleep_arrange"].map(lambda x: sum([e['mattress'] for e in x]))

x_test["sleep_arrange"] = x_test["sleep_arrange"].map(lambda x: eval(x))
x_test["sleep_arrange_count"] = x_test["sleep_arrange"].map(lambda x: len(x))
x_test["sleep_arrange_single_sum"] = x_test["sleep_arrange"].map(lambda x: sum([e['single'] for e in x]))
x_test["sleep_arrange_double_sum"] = x_test["sleep_arrange"].map(lambda x: sum([e['double'] for e in x]))
x_test["sleep_arrange_mattress_sum"] = x_test["sleep_arrange"].map(lambda x: sum([e['mattress'] for e in x]))

```

```python
df.drop(columns=["types", "regions", "rules", "features", "sleep_arrange"], inplace=True)
x_test.drop(columns=["types", "regions", "rules", "features", "sleep_arrange"], inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder
str_cols = ["allocation", "is_clean", "is_new", "is_instant", "is_plus", "province.id"]
encoder = OrdinalEncoder()
df[str_cols] = encoder.fit_transform(df[str_cols])
x_test[str_cols] = encoder.transform(x_test[str_cols])
df.select_dtypes(exclude=['number'])
```

```python
X = df.drop(columns=["min_price"])
Y = df["min_price"]
x_train,x_val,y_train,y_val = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
x_train = x_train.astype(float)
x_val = x_val.astype(float)
x_test = x_test.astype(float)
```
