```python
# TODO:
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

train = pd.read_csv('Data/train.csv')
test = pd.read_csv('Data/test.csv')

train.head()
```

    <br>
    تکنیک‌هایی که استفاده می‌کنید به شکل مستقیم مورد ارزیابی توسط سامانه داوری قرار <b>نمی‌گیرند.</b> بلکه همه آن‌ها در دقت مدل شما تاثیر خواهند گذاشت؛ بنابراین هر چه پیش‌پردازش/مهندسی ویژگی بهتری انجام دهید تا دقت مدل بهبود پیدا کند، امتیاز بیشتری از این سوال کسب خواهید کرد.
    در این قسمت شما می‌توانید بخشی از داده‌ی موجود را برای اعتبارسنجی در نظر بگیرید.
</font>
</p>

```python
import numpy as np
import pandas as pd

def feature_engineer(df, eps=1e-9):
    df = df.copy()
    df["Temp_sq"] = df["Temperature"] ** 2
    df["Pressure_sq"] = df["Pressure"] ** 2

    df["Temp_Pressure_ratio"] = df["Temperature"] / (df["Pressure"] + eps)
    df["Temp_Flow_interaction"] = df["Temperature"] * df["Flow Rate"]
    df["Vib_Rot_interaction"] = df["Vibration"] * df["Rotation Speed"]
    df["Humidity_Temp_ratio"] = df["Humidity"] / (df["Temperature"] + eps)

    numeric_cols = ["Temperature", "Pressure", "Vibration", "Flow Rate", "Humidity", "Rotation Speed"]
    means = df[numeric_cols].mean()
    stds = df[numeric_cols].std().replace(0, eps)
    for c in numeric_cols:
        df[f"{c}_z"] = (df[c] - means[c]) / stds[c]

    df["Vibration_log1p"] = np.log1p(df["Vibration"].clip(lower=0))

    df["Flow_per_Rotation"] = df["Flow Rate"] / (df["Rotation Speed"] + eps)

    sensor_cols = ["Temperature", "Pressure", "Vibration", "Flow Rate", "Humidity", "Rotation Speed"]
    df["Sensors_sum"] = df[sensor_cols].sum(axis=1)
    df["Sensors_mean"] = df[sensor_cols].mean(axis=1)

    df["Temp_roll_mean_3"] = df["Temperature"].rolling(window=3, min_periods=1).mean()
    df["Flow_roll_std_3"] = df["Flow Rate"].rolling(window=3, min_periods=1).std().fillna(0)

    # ضریب تغییرات برای Pressure و Flow نسبت به میانگین پنجره 3
    df["Pressure_cv"] = df["Pressure"] / (df["Pressure"].rolling(window=3, min_periods=1).mean() + eps)
    df["Flow_cv"] = df["Flow Rate"] / (df["Flow Rate"].rolling(window=3, min_periods=1).mean() + eps)

    return df

train = feature_engineer(train)
test  = feature_engineer(test)

print(train.head())
print(test.head())
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(train)
X_test_scaled = scaler.transform(test)
```

    حال که داده را پاکسازی کرده و احتمالا ویژگی‌هایی را به آن افزوده یا از آن حذف کرده‌اید، وقت آن است که مدلی آموزش دهید که بتواند متغیر هدف این مسئله را پیش‌بینی کند. در این قسمت استفاده از هر نوع مدل مربوط به یادگیری ماشین کلاسیک یا یادگیری عمیق مجاز می‌باشد.

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazirmatn;font-size:medium">

در این بخش، وظیفه شما توسعه مدلی است که با استفاده از داده‌های آموزش و بدون دسترسی به برچسب‌ها (labels) بتواند تشخیص دهد که هر نمونه از داده‌های آزمون، یک نمونه عادی است یا خیر.

```python
# TODO:
model = IsolationForest(n_estimators=200, contamination=0.05, random_state=42)
model.fit(X_train_scaled)

train_pred = model.predict(X_train_scaled)
test_pred = model.predict(X_test_scaled)

train_pred = np.where(train_pred == -1, 1, 0)
test_pred = np.where(test_pred == -1, 1, 0)
```
