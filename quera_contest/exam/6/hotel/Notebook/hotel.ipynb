```python
# Regression imports
import pandas as pd
import numpy as np


from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder,LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, VotingRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR


```

```python
train = pd.read_csv("../Data/train.csv",parse_dates=["date",])
x_test = pd.read_csv("../Data/test.csv",parse_dates=["date",])
```

```python
train
```

```python
train.drop(columns=["id","hotelcode","los","currency","mealinclusiontype","date"],inplace=True)
x_test.drop(columns=["id","hotelcode","los","currency","mealinclusiontype","date"],inplace=True)
```

```python
train.isnull().sum()
```

```python
train['is_block'] = (train['hotelblock'] == 'Sold out').astype(int)
x_test['is_block'] = (x_test['hotelblock'] == 'Sold out').astype(int)
train.drop(columns=["hotelblock"],inplace=True)
x_test.drop(columns=["hotelblock"],inplace=True)
```

```python
train['roomtype'] = (
    train['roomtype']
    .str.lower()
    .str.strip()
    .str.replace(r'[^a-z\s]', '', regex=True)  # remove punctuation/numbers
)
x_test['roomtype'] = (
    x_test['roomtype']
    .str.lower()
    .str.strip()
    .str.replace(r'[^a-z\s]', '', regex=True)  # remove punctuation/numbers
)

def simplify_roomtype(x):
    if 'single' in x:
        return 'Single'
    elif 'double' in x:
        return 'Double'
    elif 'twin' in x:
        return 'Twin'
    elif 'triple' in x:
        return 'Triple'
    elif 'quad' in x:
        return 'Quad'
    elif 'family' in x:
        return 'Family'
    elif 'suite' in x:
        return 'Suite'
    else:
        return 'Other'

train['roomtype'] = train['roomtype'].apply(simplify_roomtype)
x_test['roomtype'] = x_test['roomtype'].apply(simplify_roomtype)


le = LabelEncoder()
train['roomtype'] = le.fit_transform(train['roomtype'])
x_test['roomtype'] = le.transform(x_test['roomtype'])
```

```python
def simplify_ratetype(x):
    x = str(x).lower()
    if 'non-refundable' in x or 'non refundable' in x:
        return 'non_refundable'
    elif 'free cancellation' in x:
        return 'free_cancellation'
    elif 'pay at the hotel' in x:
        return 'pay_at_hotel'
    elif 'cancellation policy' in x:
        return 'cancellation_policy'
    else:
        return 'other'

train['ratetype'] = train['ratetype'].apply(simplify_ratetype)
x_test['ratetype'] = x_test['ratetype'].apply(simplify_ratetype)

Le = LabelEncoder()
train['ratetype'] = Le.fit_transform(train['ratetype'])
x_test['ratetype'] = Le.transform(x_test['ratetype'])
```

```python
train["amenities"].value_counts()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
import pandas as pd
import re

# --- Functions ---
def extract_area(text):
    if pd.isna(text):
        return None
    match = re.search(r'(\d+)\s*m²', text)
    return int(match.group(1)) if match else None


# --- Extract features for train ---
train_area_col = train["description"].apply(extract_area)
# --- Extract features for test ---
test_area_col = x_test["description"].apply(extract_area)



# --- Combine with area ---
train = pd.concat([train, pd.DataFrame({'area_m2': train_area_col})], axis=1)
x_test = pd.concat([x_test, pd.DataFrame({'area_m2': test_area_col})], axis=1)


train.drop(columns=["description"],inplace=True)
x_test.drop(columns=["description"],inplace=True)
```

```python
train["area_m2"].fillna(train["area_m2"].median(), inplace=True)
x_test["area_m2"].fillna(x_test["area_m2"].median(), inplace=True) 
```

```python
from sklearn.preprocessing import MultiLabelBinarizer



# 1️⃣ Split each row into a list of amenities
def split_amenities(text):
    if pd.isna(text) or text.strip() == "":
        return []
    # Split by semicolon and remove trailing colons and extra spaces
    return [t.strip().rstrip(":").lower() for t in text.split(';') if t.strip()]

amenity_lists = train["amenities"].apply(split_amenities)

# 2️⃣ MultiLabelBinarizer
mlb = MultiLabelBinarizer()
binary_matrix = mlb.fit_transform(amenity_lists)

test_binary_matrix = mlb.transform(x_test["amenities"].apply(split_amenities))

# 3️⃣ Convert to DataFrame
amenities_df = pd.DataFrame(binary_matrix, columns=mlb.classes_, index=train["amenities"].index)

test_amenities_df = pd.DataFrame(test_binary_matrix, columns=mlb.classes_, index=x_test["amenities"].index)

train = pd.concat([train, amenities_df], axis=1)
x_test = pd.concat([x_test, test_amenities_df], axis=1)


train.drop(columns=["amenities"],inplace=True)
x_test.drop(columns=["amenities"],inplace=True)
```
