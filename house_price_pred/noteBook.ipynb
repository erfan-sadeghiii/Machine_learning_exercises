# Pre-proccessing

 Area --> int
<br> Price --> DELETE
<br> Parking | Warehouse | Elevator --> One_Hot encode
<br> Address --> label-encode | target-encode

```python
data = data.drop("Price",axis=1)
```

```python
data = data.dropna()
cols_bool = ['Parking', 'Warehouse', 'Elevator']
data[cols_bool] = data[cols_bool].astype(int)
```

```python
data[data['Area'].str.replace(',', '').str.isnumeric() == False]['Area']
data = data.dropna(subset=['Area'])
```

```python
import re

def clean_area(x):
    if pd.isna(x):
        return None
    x = str(x)
    x = x.replace(',', '')      # remove commas
    x = re.sub(r'\D', '', x)    # keep only digits
    return int(x) if x.isdigit() else None

data['Area'] = data['Area'].apply(clean_area)


data['Area'] = data['Area'].astype(int)
```

```python
data['Area'].describe()
```

### Area outlier handled logically

```python
data = data[(data['Area'] >= 20) & (data['Area'] <= 2000)]
```

```python
le = LabelEncoder()
data["Address"] = le.fit_transform(data["Address"])


# import pandas as pd
# from sklearn.preprocessing import LabelEncoder

# counts = data["Address"].value_counts()

# rare_labels = counts[counts < 60].index

# data["Address"] = data["Address"].replace(rare_labels, "other")

# le = LabelEncoder()
# data["Address"] = le.fit_transform(data["Address"])

# data["Address"].value_counts()
```

```python
g = data.groupby("Address")["Price(USD)"]

data["addr_price_mean"] = data["Address"].map(g.mean())
data["addr_price_min"]  = data["Address"].map(g.min())
data["addr_price_max"]  = data["Address"].map(g.max())
```

```python
def estimate_floors(row):
    area = row["Area"]
    rooms = row["Room"]
    elevator = row["Elevator"]  # 0 یا 1
    
    if rooms == 0:
        return 1

    ratio = area / rooms  # نسبت مساحت به اتاق

    # منطق پایه
    if ratio < 25:
        floor = 1
    elif ratio < 55:
        floor = 1
    elif ratio < 90:
        floor = 2
    else:
        floor = 3

    # تاثیر آسانسور
    if elevator == 0 and floor > 2:
        floor = 2  # اگر آسانسور نیست، حداکثر 2 طبقه
    elif elevator == 1 and floor == 3:
        floor = 3  # با آسانسور اجازه طبقه بالاتر

    return floor

data["FloorCount"] = data.apply(estimate_floors, axis=1)
```

```python
data["AreaPerRoom"] = data["Area"] / data["Room"].replace(0, 1)
data["AreaPerFloor"] = data["Area"] / data["FloorCount"].replace(0, 1)
data["ParkingPerRoom"] = data["Parking"] / data["Room"].replace(0,1)
data["FloorElevator"] = data["FloorCount"] * data["Elevator"]
data["AreaFloorElevator"] = data["AreaPerFloor"] * data["Elevator"]
data["LargeArea"] = (data["Area"] > data["Area"].median()).astype(int)
data["LuxuryUnit"] = ((data["Parking"]>0) & (data["Elevator"]==1) & (data["Room"]>=3)).astype(int)
data["PricePerArea"] = data["Price(USD)"] / data["Area"].replace(0,1)
data["PricePerRoom"] = data["Price(USD)"] / data["Room"].replace(0,1)
freq = data["Address"].value_counts()
data["AddressFreq"] = data["Address"].map(freq)
```

# splitting data

```python
X = data[['Area', 'Room', 'Parking', 'Elevator', 'Warehouse', 'Address',"addr_price_mean","AddressFreq","PricePerRoom","PricePerArea","LuxuryUnit","LargeArea","AreaFloorElevator","FloorElevator","ParkingPerRoom","AreaPerFloor","FloorCount","AreaPerRoom"	,"addr_price_min"	,"addr_price_max"]]
y = data['Price(USD)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 1️⃣ Create the scaler
scaler = StandardScaler()

# 2️⃣ Fit only on the training set, then transform both
train_scaled = scaler.fit_transform(X_train)

test_scaled = scaler.transform(X_test) # our test is x_test
```

# best model and parameter

```python
import optuna
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score


def objective(trial):
    model_name = trial.suggest_categorical("model", ["svr", "linear", "dt", "xgb"])

    if model_name == "svr":
        model = SVR(
            C=trial.suggest_float("C", 0.1, 100.0, log=True),
            epsilon=trial.suggest_float("epsilon", 0.01, 1.0),
            kernel="rbf",
        )

    elif model_name == "linear":
        model = LinearRegression()

    elif model_name == "dt":
        model = DecisionTreeRegressor(
            max_depth=trial.suggest_int("max_depth", 2, 20),
            min_samples_split=trial.suggest_int("min_samples_split", 2, 20),
            random_state=42,
        )

    else:  # xgb
        model = XGBRegressor(
            n_estimators=trial.suggest_int("n_estimators", 100, 800),
            learning_rate=trial.suggest_float("learning_rate", 0.01, 0.3),
            max_depth=trial.suggest_int("max_depth", 3, 12),
            objective="reg:squarederror",
            random_state=42,
        )

    # Train
    model.fit(train_scaled, y_train)

    # Predict
    y_pred_train = model.predict(train_scaled)
    y_pred_test = model.predict(test_scaled)

    # Scores
    r2_train = r2_score(y_train, y_pred_train)
    r2_test = r2_score(y_test, y_pred_test)

    # Save scores into trial metadata
    trial.set_user_attr("r2_train", r2_train)
    trial.set_user_attr("r2_test", r2_test)

    return r2_test   # optimize only test score


# Run Optuna
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=50)

# Best trial reference
best_trial = study.best_trial

# Clean output
print("\n===== Best Result =====")
print(f"R² Train: {best_trial.user_attrs['r2_train']:.4f}")
print(f"R² Test : {best_trial.user_attrs['r2_test']:.4f}")

print("\n===== Best Parameters =====")
for k, v in best_trial.params.items():
    print(f"{k}: {v}")
```
